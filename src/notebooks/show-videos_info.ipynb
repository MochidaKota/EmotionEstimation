```python
import pandas as pd
import plotly.graph_objects as go
df = pd.read_csv("/mnt/iot-qnap3/mochida/medical-care/emotionestimation/data/labels/PIMD_A/label_video1-25.csv")
```

```python
videos_info = {
    'video_name':[],
    'length':[],
    '0':[],
    '1':[],
    '2':[]
}

for i in range(1,26):
    video_name = 'video' + str(i)
    df_video = df[df['video_name'] == video_name]
    videos_info['video_name'].append(video_name)
    videos_info['length'].append(len(df_video))
    videos_info['0'].append(len(df_video[df_video['label'] == 0]))
    videos_info['1'].append(len(df_video[df_video['label'] == 1]))
    videos_info['2'].append(len(df_video[df_video['label'] == 2]))

videos_info = pd.DataFrame(videos_info)

# add row of total
# except video_name
videos_info.loc['total'] = videos_info.iloc[:,1:].sum()
# all values to int
videos_info = pd.concat([videos_info.iloc[:,0], videos_info.iloc[:,1:].astype(int)], axis=1)
videos_info
```

```python
def show_label_distribution(label_df, video_name):
    # fig size = (width, height)
    fig = go.Figure(layout=go.Layout(width=700, height=200))
    video_df = label_df[label_df['video_name'] == video_name]
    video_df = video_df.sort_values('frame')
    # if label == 1, color = red, label == 2, color = green
    fig.add_trace(go.Bar(x=video_df['frame'], y=video_df['label'], marker_color=video_df['label'].map({0:'blue', 1:'red', 2:'green'})))
    fig.update_yaxes(range=[0, 1], showticklabels=False)
    fig.update_layout(
        title=video_name,
        xaxis_title="frame",
        font=dict(
            size=18,
        ),
        margin=dict(
            l=0, 
            r=0, 
            t=40, 
            b=0
        ) 
    )
    # tyght layout
    fig.update_layout(margin=dict(l=0, r=0, t=45, b=0))
    fig.show()
```
