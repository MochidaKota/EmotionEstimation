```python
from utils import util 
from PIL import Image
from IPython.display import Image as IPImage, display
import os
```

```python
def get_sequence_imgs(head_img_path, window_size):
    imgs = []
    paths = util.get_sequence_img_path(current_img_path=head_img_path, window_size=window_size, current_position='head')
    video_name, frame_num = util.get_video_name_and_frame_num(head_img_path)
    for path in paths:
        imgs.append(Image.open(path))
    
    gif_path = f'{video_name}_{str(frame_num).zfill(4)}.gif'    
    # make gif
    imgs[0].save(gif_path, save_all=True, append_images=imgs[1:], duration=50, loop=0)
    
    display(IPImage(gif_path, width=400, height=400))
    
    # remove gif
    os.remove(gif_path)
    
```

```python
video_name = 'video19'
frame_num = 1450

img_path = f'/mnt/iot-qnap3/mochida/medical-care/emotionestimation/data/interim/PIMD_A/{video_name}/{video_name}_{str(frame_num).zfill(4)}.jpg'
get_sequence_imgs(img_path, window_size=50)
```
